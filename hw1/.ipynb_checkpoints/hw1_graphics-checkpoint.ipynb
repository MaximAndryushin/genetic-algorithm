```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def func(x) :
    return np.cos(np.exp(x)) / np.sin(np.log(x))
x = np.linspace(2,4,1 << 20)
y = func(x)
```

```python
max_y = max(y)
best_x = x[np.where(y == max_y)]

print("best result(maximum of func):", max_y, "\n", "x = ", best_x)
```

График функции с выделенной точкой максимума

```python
plt.subplots(figsize=(10,6))
plt.plot(x, y)
plt.plot(best_x, max_y, 'r+', markersize=12)
plt.xlabel('x')
plt.ylabel('FUNCTION')
plt.title('cos(exp(x)) / sin(log(x))')
plt.grid()
```
