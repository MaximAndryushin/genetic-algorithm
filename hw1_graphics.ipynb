```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,4,1 << 20)
y = np.cos(np.exp(x)) / np.sin(np.log(x))
```

```python
plt.subplots(figsize=(10,6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('FUNCTION')
plt.title('cos(exp(x)) / sin(log(x))')
plt.grid()
```
