# LAB1 Knapsack problem

```python
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
```

## График зависимости времени нахождения одного решения методом полного перебора от amax

```python
amax = [144715, 360974, 1048576, 3697661, 16777216, 106528681, 1073741824]
time = [0.183578, 0.275643, 0.370224, 0.612793, 0.807996, 0.813604, 0.710434]

plt.subplots(figsize=(10,6))
plt.plot(amax, time)
plt.xlabel('amax')
plt.xscale('log')
plt.ylabel('time of work')
plt.grid()
```

С интерполяцией

```python
new_x = np.linspace(min(amax), max(amax), 1000)
a_BSpline = make_interp_spline(amax, time)
new_y = a_BSpline(new_x)

plt.subplots(figsize=(10,6))
plt.plot(new_x, new_y)
plt.xlabel('amax')
plt.xscale('log')
plt.ylabel('time of work')
plt.grid()
```
